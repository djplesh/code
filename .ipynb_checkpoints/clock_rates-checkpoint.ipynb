```python
import tetra_analysis.timestamps.ctr_to_timestamp4 as get
import datetime as dt
```

```python
%matplotlib inline

d = dt.date(2018,3,1)
box = 'LSU_01'
devices = ['Dev1', 'Dev2']

clk_avgs = {'Dev1':[], 'Dev2':[]}
plot_color = {'Dev1': 'b', 'Dev2': 'r'}

for dev in devices:
    path0 = 'E:/' + box + '/' + d.strftime("%Y_%m") + '/' + dev + '/'


    all_clks = []
    file_list = get.day_files(path0, 'ctr1', d)
    for f in file_list:
        pps_idx, pps_ctr1, clks = get.get_pps(f)
        for c in clks:
            all_clks.append(c)
        if len(np.array(clks) > 0):
            clk_avgs[dev].append(np.average(clks))
    plt.figure()
    plt.plot(all_clks, lw = 0, marker = '.', color = plot_color[dev])
    plt.show()
    plt.figure()
    plt.plot(clk_avgs[dev], lw = 0, marker = '.', color = plot_color[dev])
    plt.show()
```
